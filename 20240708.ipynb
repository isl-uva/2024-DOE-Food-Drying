```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the constants and formulas
def k(T):
    return 0.0008 * np.exp(0.0485 * T)

def Me(T):
    return 0.317 - 0.0053 * T

def M(t, Mo, T):
    ke = k(T)
    Me_val = Me(T)
    return (Mo - Me_val) * np.exp(-ke * t) + Me_val

# Set the parameters for the plot
T1 = 36  # Example temperature value, you can change this
T2 = 48
T3 = 57
Mo = 1.0  # Initial moisture content, example value
time = np.linspace(0, 1000, 1000)  # Time range from 0 to 100

# Compute M(t) for the given T and time
moisture_content1 = 100*M(time, Mo, T1)
moisture_content2 = 100*M(time, Mo, T2)
moisture_content3 = 100*M(time, Mo, T3)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(time, moisture_content1, label=f'T={T1}C')
plt.plot(time, moisture_content2, label=f'T={T2}C')
plt.plot(time, moisture_content3, label=f'T={T3}C')
plt.xlabel('Time (min)')
plt.ylabel('Moisture Content (%)')
plt.title('Moisture Content Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
